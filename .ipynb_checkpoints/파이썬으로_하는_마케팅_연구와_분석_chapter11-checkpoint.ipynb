```python
# 분류 프로젝트에는 일반적으로 최소한 다음 단계가 포함된다.
# 1. 각 관측치에 대한 그룹 구성원이 알려져 있거나 이미 할당된 데이터셋이 수집된다.
# 2. 데이터셋은 훈련 집합과 테스트 집합으로 나뉜다. 일반적인 패터는 관측치의 50~80%(대개 70%)를 훈련집합으로 선택하고
#    나머지는 테스트 집합에 할당한다.
# 3. 훈련 데이터에 있는 구성원을 예측하는 것을 목표로 예측 모델이 구축된다.
# 4. 결과 모델은 테스트 데이터를 사용해 성능을 평가한다. 
#    성능이 우연성(기본율)을 초과하는지 확인하기 위해 평가된다. 또한 방법이 합리적인 대안 (그리고 더 간단하거나 잘 알려진)
#    모델보다 더 나은 성능을 발휘하는지 평가할 수 있다.
```

## 나이브 베이즈 분류: GaussianNB()

```python
# 훈련 데이터를 사용해 독립이라고 간주되는 각 예측 변수의 함수로서 부류 구성원의 확률을 학습
# 새로운 데이터에 적용할 때 부류 멤버십은 예측자의 조합에 의해 할당된 결합 확률에 따라 가능성이 높은 범주로 할당

import pandas as pd
seg_df = pd.read_csv('http://bit.ly/PMR-ch5')
seg_df['is_female'] = seg_df.gender == 'female'
seg_sub = seg_df.drop(['Segment', 'gender'], axis=1)
seg_sub.head()
```

```python
# 1. 분류기를 학습하는 첫 번째 단계는 데이터를 훈련 데이터와 테스트 데이터로 분류하는 것
# 그러면 모델이 테스트에서 작동하는 (또는 대신 훈련 데이터에 과적화되는지) 확인할 수 있다.
# 훈련에 사용할 데이터의 70%를 선택하고 선택하지 않은 부분을 홀드아웃(테스트) 데이터로 유지한다.

# 분류에는 새 값을 할당하는 방법을 배우기위해 알려진 세그먼트 할당 값이 필요한데,
# 여기서는 seg_labels에 저장할 것이다. 
# 일반적 규칙은 독립변수는 x에 할당하고 종속 변수(레이블)은 y에 할당한다.
```

```python
import numpy as np

seg_labels = seg_df.Segment
np.random.seed(537)
rand_idx = np.random.rand(seg_labels.shape[0])
train_idx = rand_idx <= 0.7
test_idx = rand_idx > 0.7

X_train = seg_sub.iloc[train_idx]
X_test = seg_sub.iloc[test_idx]

y_train = seg_labels.iloc[train_idx]
y_test = seg_labels.iloc[test_idx]
```

```python
from sklearn import naive_bayes

nb = naive_bayes.GaussianNB()

nb.fit(X_train, y_train)

list(zip(nb.classes_, nb.class_prior_))
```

```python
# class_prior_값을 보면 모델 작동 방식에 대한 통찰을 얻을 수 있다.
# 1. 세그먼트 멤버십의 사전 우도(즉, 다른 정보가 추가되기 전에 예상되는 멤버십 확률)은 Moving up에서 27.1%
#                                                                   Suburb mix는 32.2% 
# 모델은 각 예측 변수에 조건부 확률을 사용한다.

# NB 분류기는 훈련 데이터에서 찾은 세그먼트를 조건으로 성별, 연령 등의 관찰된 확률로 시작한디.
# 그런 다음 베이즈의 규칙을 사용해 성별, 연령 등에 따라 세그먼트 확률을 계산
# 그런 다음 테스트 데이터와 같은 새로운 관측치에서 세그먼트 멤버십을 추정하는데 사용할 수 있다.

# 훈련 데이터와 테스트 데이터 모두에 대해 예측을 생성할 수 있으며 일부 사용자에 대한 참 값과 예측 레이블을 확인해볼 수 있음.

predictions = nb.predict(seg_sub)
seg_sub_pred = seg_sub.copy()
seg_sub_pred['predition'] = predictions
seg_sub_pred['true_segment'] = seg_df['Segment']
seg_sub_pred.sample(5)
```

```python
# 전반적으로 얼마나 잘 수행되었는가?
# 테스트 데이터에 score() 메서드를 사용

# 예측된 세그먼트 멤버십과 참 세그먼트 멤버십 간에 일치된 값인 정확도 점수를 반영

nb.score(X_test, y_test)
```

```python
# 원시 정확도 대신 예측력 측면에서 모델의 성능을 평가해야 한다.
# 일반적으로 사용되는 측정 항목은 정밀도와 재현율의 조화 평균이 F 점수이다.

from sklearn import metrics

y_pred = nb.predict(X_test)
metrics.f1_score(y_true=y_test, y_pred=y_pred, average='weighted')
```

```python
# average 매개변수는 4개의 다른 세그먼트(sklearn에서는 클래스라고 함)의 성능을 결합하는 방법을 지정한다 (여기서는 가중치 적용)
# 즉 F 점수는 각 클래스에 대해 계산되고 평균은 모집단의 클래스 비율에 따라 가중된다.

# 머신러닝에서 혼동 행렬(confusion matrix)로 알려진 것을 사용해 각 범주의 성능을 비교하기
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def confusion_matrix(y_true, y_pred, model):
    conf_mat = metrics.confusion_matrix(y_true, y_pred)
    
    sns.heatmap(
        conf_mat.T,
        xticklabels=model.classes_,
        yticklabels=model.classes_,
        annot=True, fmt='d'
    )
    
    print(conf_mat.T)
    plt.xlabel('true label')
    plt.ylabel('predicted label')
    
confusion_matrix(y_test, y_pred, nb)

# 올바른 예측은 대각에 표시됨
# NB 예측은 moving_up을 제외하고 각 세그먼트에서 대부분의 관측치에 대해 정확했음.
# 
```

```python
# 이 점은 예측의 비대칭성을 보여줌.
# 모델은 참 긍정와 참 부정을 모두 정확하게 식별해야 함.

# 머신러닝의 중요한 통계 개념에 해당하는 이러한 요구사항 사이에는 상호 보완성이 있음.
# 1. 정밀도로, 특정 레이블이 참 레이블로 식별되는 샘플의 비율.
#    모든 긍정 중 참 긍정이 레이블인 비율
#    (정밀도 = predicted label (대각 부분) / true label (가로의 전체 개수))
#     정밀도 = 참긍정 / (참긍정 + 거짓 긍정)
# 2. 두번째 중요한 개념은 재현율 또한 민감도이며 참 긍정 비율
#    이는 모든 긍정 비율 중 제대로 식별된 비중.
#    재현율 = 참긍정 / (참긍정 + 거짓 부정)
#    재현율 = 대각선 값 / predicted label 전체 값

# 3. F 점수
#    F = 2 * (정밀도 * 재현율)/(정밀도 + 재현율)
# 함수로 구현하기

def return_precision_recall(y_true, y_pred, model):
    conf_mat = metrics.confusion_matrix(y_true, y_pred)
    
    precision = pd.Series(metrics.precision_score(y_test,
                                                  y_pred, 
                                                  average=None),
                          index=model.classes_)
    recall = pd.Series(metrics.recall_score(y_test,
                                            y_pred,
                                            average=None),
                       index=model.classes_)
    f1 = pd.Series(2*(precision * recall)/(precision + recall),
                   index = model.classes_)
    
    return pd.DataFrame([precision, recall, f1], index=['precision','recall', 'f1'])

return_precision_recall(y_test, y_pred, nb)
```

```python
# PCA 공간에서 결정 경계를 시각화하기

from sklearn import clone, decomposition

def plot_decision_pca(model, X, y):
    width, height = 500,500
    
    # PCA를 사용해 x값을 변환
    p = decomposition.PCA(random_state=132, svd_solver='full')
    X_transformed = p.fit_transform(X.iloc[:,:2])
    
    # 첫 2개의 차원을 끄집어낸다.
    x0 = X_transformed[:, 0]
    x1 = X_transformed[:, 1]
    
    # min, max 값 사이의 균등 분포 값을 얻는다.
    x0_g = np.linspace(x0.min(), x0.max(), width)
    x1_g = np.linspace(x1.min(), x1.max(), height)
    
    # 각 벡터로부터 균등 분포된 '그리드(grid)'를 얻는다
    xx, yy = np.meshgrid(x0_g, x1_g)
    
    # 샘플 값을 같이 쌓는다
    X_grid_transformed = np.vstack([xx.ravel(), yy.ravel()]).T
    
    # 비PCA 변환값을 얻기 위해 역 변환을 한다.
    X_grid = p.inverse_transform(X_grid_transformed)
    
    # 첫 두 PCA 열로부터 역 변환된 것을 이용해 모델 복제를 적합화한다.
    # 표본 값을 예측한다
    model_c = clone(model)
    model_c.fit(p.inverse_transform(np.vstack([x0,x1]).T), y)
    X_grid_labels = model_c.predict(X_grid)
    
    # 클래스 매퍼(mapper)를 생성해 클래스 문자열을 정수로 매핑한다.
    class_mapper = {class_:i for i, class_ in enumerate(model.classes_)}
    
    plt.figure(figsize=(6,6))
    # 예측 값을 도식화
    
    a = plt.scatter(
        x0, x1, 
        c=[class_mapper[label] for label in y],
        cmap = plt.cm.rainbow, edgecolor='k',
        vmin=0, vmax=3
    )
    
    plt.contourf(
        xx, yy,
        np.reshape([class_mapper[label] for label in X_grid_labels], (width,height)),
        cmap=a.cmap, alpha=.5, levels=3
    )
    
    cb = plt.colorbar(ticks=[.5, 1.2, 2, 2.8])
    _ = cb.ax.set_yticklabels(model.classes_)
    plt.title('Decision boundaries with true values overlaid')
    plt.xlabel('First principal component')
    plt.ylabel('Second principal component')
    
plot_decision_pca(nb, X_test, y_test)
```

```python
# 간단히 요약하면, 
# 이 작업은 PCA 공간의 처음 두 성분 내의 그리드에서 균등하게 샘플링해 각 지점에서 모델 예측을 평가하는 것
# 그럼 다음 테스트 집합의 참 값을 중첩하고 모델 예측과 일치하지 않는 부분을 확인할 수 있음.
```

```python
# 이 경우 suburb mix와 moving up 세그먼트가 산재돼 있음을 알 수 있음.
# 이는 혼동 행렬에서 본 것과 일치한다.
# 이를 어떻게 처리할지에 대한 질문이 바로 비즈니스 결정이 된다.
# 예를 들어, 이 두 세그먼트를 구분하고 축소할 필요가 없다고 결정할 수 있다.
# 또는 교육 수준이나 자동차 소유와 같이 이 두 세그먼트에서 서로 다른 유형의 데이터를 수집하려고 할 수 있다.
```

## 랜덤 포레스트 분류

```python
# 단일 모델을 데이터에 적합화하지 않고 그 대신 데이터를 연합으로 분류하는 모델의 앙상블을 구축
# RF 는 많은 수의 분류 트리를 적합화해 이를 수행함.
# 다양한 모델을 찾기 위해 각 트리는 일부 예측 변수만 사용해 일부 관측치(이 경우 고객)에만 적합하도록 최적화함.
# 모든 트리의 앙상블은 포레스트가 된다.


# 새로운 케이스가 예측될 때면, 모든 트리에서 예측하고 최종 결정은 가장 많은 표를 얻은 합의 값에 부여됨.
# 이러한 방식으로 랜덤 포레스트는 정말한 모델 사양에 대한 종속성을 피하면서
# 공선성 혹은 와이드 데이터와 같은 어려운 데이터 조건에서도 탄력성을 유지.
```

```python
# PCA 공간에서 의사 결정 곤간을 검사하면 모델 약점 영역을 이해하는 데 도움이 될 수 있지만,
# 실제 특징을 기반으로 모델 의사 결정을 주도하는 요인에 대해서는 많은 통찰력을 제공하지 않는다.
# 또한 연령과 소득 같은 특징 쌍에 대한 결정 경계를 검사할 수 있다.
```

```python
def pairwise_decision_boundary(model, X_train, y_train,
                               X_test, y_test,
                               first_column, second_column,
                               jitter=False):
    
    width,height =100,100
    # 문자 클래스와 정수를 매팡하기 위한 클래스 매퍼를 생성
    class_mapper = {c:i for i,c in enumerate(model.classes_)}
    
    x0 = X_train[first_column]
    x1 = X_train[second_column]
    # 최대 최소 값 사이에 균등 분포된 값을 얻음
    x0_g = np.linspace(x0.min(), x0.max(), width)
    x1_g = np.linspace(x1.min(), x1.max(), height)
    # 각 벡터에서 균등 분포한 '그리드' 생성
    xx,yy = np.meshgrid(x0_g, x1_g)
    # 모든 샘플 값을 같이 스택(stack)함
    X_grid = np.vstack([xx.ravel(), yy.ravel()]).T
    
    model_c = clone(model)
    model_c.fit(X_train.loc[:, [first_column, second_column]], y_train)
    X_grid_labels = model_c.predict(X_grid)
    
    # 예측값 도식화
    # 구독상태와 같은 불연속이나 부울 값을 보려면 jitter 같은 선택인수가 있음. (그렇지 않으면 포인트가 모두 서로 겹쳐 보임.)
    j_x0, j_x1 = 0,0
    if jitter:
        j_x0 = (np.random.random(X_test.shape[0])-.5)/10
        j_x1 = (np.random.random(X_test.shape[0])-.5)/10
    
    a = plt.scatter(
        X_test[first_column] + j_x0,
        X_test[second_column] + j_x1,
        c = [class_mapper[l] for l in y_test],
        cmap=plt.cm.rainbow,
        edgecolor='k', vmin=0, vmax=3
    )
    plt.contourf(
        xx, yy, 
        np.reshape([class_mapper[l] for l in X_grid_labels],
                   (width, height)),
        cmap=a.cmap, alpha=.5, levels=3
    )
    plt.title('Decision boundaries with true values overlaid')
    plt.xlabel(first_column)
    plt.ylabel(second_column)
    cb = plt.colorbar(ticks=[.5, 1.2, 2, 2.8])
    cb.ax.set_yticklabels(model.classes_)

pairwise_decision_boundary(rf, X_train, y_train, X_test, y_test, 'age', 'income')
```

## 랜덤 포레스트 변수 중요도

```python
# 랜덤 포레스트는 특히 하나의 일반적인 마케팅 문제, 즉 분류 변수의 중요도를 평가하는데 유용하다.
# RF는 많은 트리를 적합화하며, 각 트리는 데이터의 일부에 최적화돼있다.
# 트리의 성능을 좀 더 일반적으로 평가하기 위해 나머지 데이터(out of bag)을 사용한다.
# 각 트리는 변수의 하위 집합만 사용하므로 RF 모델은 관측치보다 더 많은 (심지어 훨씬 더 많은) 예측 변수가 있는
# 매우 와이드 한 데이터를 처리할 수 있다.
```

```python
# RF 모델은 간단하면서도 강력한 방법으로 변수의 중요도를 평가한다. 
# 한 번에 하나의 변수에 대해 변수 값을 무작위로 순열(변경)하고 순열된 값을 사용해 OOB 데이터의 모델 정확도를 계산한 다음,
# 이를 실제 데이터와의 정확도와 비교한다.

# 변수가 중요한 경우 관찰된 값이 무작위로 치환되면 성능이 저하된다.
# 그러나 모델이 실제 데이터만큼 정확하다면 해당 변수는 그다지 중요하지 않다.

# 앞서 언급했듯이 모델의 목표가 변수 중요도를 결정하는 것이면 더 많은 트리를 사용하는 것이 일반적이다.
# 전체 모델이 변수 공간을 더 많이 포함하고 변수 중요도 값은 더 높은 정밀도를 갖기 때문이다.
```

```python
# RF 모델의 feature_importances_ 매개변수에서 각 특징의 계산된 중요도를 볼 수 있다.

pd.Series(rf.feature_importances_,
          index=seg_sub.columns).sort_values(ascending=False)
```

```python
# 변수 중요도를 이해하면 부류 간의 차이를 정의하는 특징을 더 깊이 이해할 수 있으므로
# 좀 더 지능적인 비즈니스 의사 결정이 가능.
```

```python
# 오류율은 처음에는 전반적으로 90%로 좋게 보일 수 있지만,
# 재현율이 단 10%에 불과하다.

# 이는 클러스터에서 클래스의 산포를 고려할 때 예상된 것.

# 한 범주가 데이터를 지배하면 다른 그룹을 예측하는 방법을 학습하기가 매우 어려움.
# 이는 어느 제품을 구매할 개인이 드문 경우, 희귀 질병이 있는 사람, 보안 위협이 되는 사람 등을
# 예측하는 것과 같은 작은 비율의 문제에서 자주 발생.
```

```python
# 일반적인 해법은 소그룹에서 더 많이 샘플링해 부류의 균형을 맞추는 것
# RF 모델에서 각 트리에 적합화할 데이터를 샘플링할 때 class_weight = 'balanced_subsample'을 수행해
# 분류기에 균형된 그룹을 사용하도록 지시할 수 있다.

# 그러나 변수 자체의 열약한 예측력을 극복하는 것만으로는 충분하지 않았다.
```

```python
# 모든 모델 매개변수에 기본 설정값을 사용했다.
# 머신러닝에서 중요한 개념 중 하나는 초매개변수 튜닝이다.

# 여기서 모델 매개변수 공간을 탐색해 최적으로 이어지는 매개변수를 식별한다.
# 최적을 찾기 위해 여러 매개변수 조합을 샘플링하는 그리드 검색을 수행할 수 있다.

# 이를 위해서는 점수 함수가 필요.
# F1 점수는 정밀도와 재현율이 균형을 이루므로 유용함.

from sklearn import model_selection

rf_sub_cv = ensemble.RandomForestClassifier(random_state=34,
                                            class_weight='balanced_subsample')
parameters = {
    'n_estimators':[10,100,500],
    'max_depth':[5,10,30],
    'min_samples_split':[2,5],
    'min_samples_leaf':[1,2,5]
}
clf = model_selection.GridSearchCV(rf_sub_cv, parameters, cv=5, scoring='f1_weighted')
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
# 초매개변수 튜닝 모델은 더 나은 성능을 보이지 않았다.

y_pred_be = clf.best_estimator_.predict(X_test)
confusion_matrix(y_test, y_pred_be, clf.best_estimator_)
```

```python
# 다른 점수 함수 시도?
# 재현율을 최적화하고 싶다고 가정해보기
# 즉, 가능한 한 잠재적인 긍정을 찾아서 거짓 긍정이 될 수 있는 것을 받아들임.
# 재현율을 점수 함수로 사용해보자

rf_sub_cv = ensemble.RandomForestClassifier(random_state=34,
                                            class_weight='balanced_subsample')
parameters = {
    'n_estimators':[10,100,500],
    'max_depth':[5,10,30],
    'min_samples_split':[2,5],
    'min_samples_leaf':[1,2,5]
}
clf = model_selection.GridSearchCV(rf_sub_cv, parameters, cv=5, scoring='recall')
clf.fit(X_train, y_train)
```
